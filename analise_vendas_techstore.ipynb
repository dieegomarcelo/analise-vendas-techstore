# Passo 1: Importar bibliotecas e carregar os dados

```python
import pandas as pd
```

```python
# Carregar o arquivo CSV
df = pd.read_csv("/content/vendas.csv")
```

```python
# Exibir as 5 primeiras linhas
print("Primeiras 5 linhas:")
print(df.head())
```

```python
# Exibir o número total de registros
print("\nNúmero total de registros:", len(df))
```

# Passo 2: Calcular a receita total

```python
# Criar coluna de receita
df["receita"] = df["quantidade"] * df["preco_unitario"]
```

```python
# Receita total
receita_total = df["receita"].sum()
print("\nReceita total:", receita_total)
```

# Passo 3: Consultas específicas
## 3.1 Filtrar vendas da categoria "Eletrônicos"

```python
eletronicos = df[df["categoria"] == "Eletrônicos"]
print("\nVendas da categoria Eletrônicos:")
print(eletronicos)
```

## 3.2 Produto mais vendido (em quantidade)

```python
produto_mais_vendido = df.groupby("produto")["quantidade"].sum().idxmax()
print("\nProduto mais vendido:", produto_mais_vendido)
```

```python
quantidade_vendida = df.groupby("produto")["quantidade"].sum().max()
print("Quantidade vendida:", quantidade_vendida)
```

## 3.3 Região com maior valor de compras

```python
regiao_top = df.groupby("regiao")["receita"].sum().idxmax()
print("\nRegião com maior valor de compras:", regiao_top)
```

```python
valor_regiao_top = df.groupby("regiao")["receita"].sum().max()
print("Valor total:", valor_regiao_top)
```

# Passo 4: Visualizações simples

```python
import matplotlib.pyplot as plt
```

```python
# Receita por categoria
df.groupby("categoria")["receita"].sum().plot(kind="bar", title="Receita por Categoria")
plt.show()
```

```python
# Receita por região
df.groupby("regiao")["receita"].sum().plot(kind="pie", autopct="%1.1f%%", title="Receita por Região")
plt.show()
```
